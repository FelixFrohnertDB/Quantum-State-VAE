```python
import numpy as np
import qutip as qt
import matplotlib.pyplot as plt
from qiskit.quantum_info import Statevector, DensityMatrix, entropy, partial_trace
from qiskit import QuantumRegister, ClassicalRegister
from qiskit import QuantumCircuit, execute
```

```python
def circuit_w(theta):
    q = QuantumRegister(4,'q')
    circuit = QuantumCircuit(q)
    circuit.ry(theta,q[0])
    
    circuit.cnot(q[0],q[1])
    circuit.cnot(q[0],q[2])
    circuit.cnot(q[0],q[3])
    return circuit

circuit_w(np.pi).draw()
```

```python
def circuit_w(theta):
    q = QuantumRegister(3,'q')
    circuit = QuantumCircuit(q)
    circuit.ry(theta,q[0])
    circuit.ch(q[0],q[1])
    circuit.cnot(q[1],q[2])
    circuit.cnot(q[0],q[1])
    circuit.x(q[0])
    return circuit

circuit_w(np.pi).draw()
```

```python
def circuit_ghz(theta):
    q = QuantumRegister(3,'q')
    circuit = QuantumCircuit(q)
    circuit.ry(theta,q[0])
    circuit.cnot(q[0],q[1])
    circuit.cnot(q[0],q[2])
    return circuit

def get_dm_ghz(theta):
    state = Statevector.from_int(0, 2**3)
    state = state.evolve(circuit_ghz(theta))
    return DensityMatrix(state)

def circuit_w(theta):
    q = QuantumRegister(3,'q')
    circuit = QuantumCircuit(q)
    circuit.ry(theta,q[0])
    circuit.ch(q[0],q[1])
    circuit.cnot(q[1],q[2])
    circuit.cnot(q[0],q[1])
    circuit.x(q[0])
    return circuit

def get_dm_w(theta):
    state = Statevector.from_int(0, 2**3)
    state = state.evolve(circuit_w(theta))
    return DensityMatrix(state)

def get_u(n):
    z = np.random.randn(n,n)/np.sqrt(2.0)
    q,r = np.linalg.qr(z)
    d = np.diagonal(r)
    ph = d/np.absolute(d)
    q = np.multiply(q,ph,q)
    return q

def apply_u(dm):
    UaUb = qt.Qobj(
        np.kron(
            np.kron(get_u(2),get_u(2)),get_u(2) )
    ,dims = [[2, 2, 2], [2, 2, 2]])
    return  UaUb * dm * UaUb.dag() 
```

```python
n_points = 101
scan_arr = np.linspace(0,np.pi/2,n_points)
cnt = 0 
rho_arr_noisy = []
rho_arr = []

for i, theta in enumerate(scan_arr):   
        
    dm = get_dm_ghz(theta) 

    rho_arr.append(qt.Qobj(dm.data,dims=[[2,2,2],[2,2,2]])) 
    rho_arr_noisy.append(apply_u( qt.Qobj(dm.data,dims=[[2,2,2],[2,2,2]]))) 
    
c_01 = []
n_01 = []

c_02 = []
n_02 = []

c_12 = []
n_12 = []


for state in rho_arr_noisy:
    c_01.append(qt.entropy_vn(state.ptrace([0,1])))
    n_01.append(qt.negativity(state,[0,1]))
    
    c_02.append(qt.entropy_vn(state.ptrace([2])))
    n_02.append(qt.negativity(state,[0,2]))
    
    c_12.append(qt.entropy_vn(state.ptrace([1,2])))
    n_12.append(qt.negativity(state,[1,2]))
    
fig, axs = plt.subplots(3,2)
axs[0,0].plot(c_01)
axs[0,1].plot(n_01)

axs[1,0].plot(c_02)
axs[1,1].plot(n_02)

axs[2,0].plot(c_12)
axs[2,1].plot(n_12)

plt.tight_layout()
```

```python
np.save("test_arr.npy",np.array([rho_.ptrace([0,1]) for rho_ in rho_arr]).real)
```

```python
n_points = 101
max_angle = 2*np.arccos(1/np.sqrt(3)) 
scan_arr = np.linspace(0,max_angle,n_points)
cnt = 0 
rho_arr_noisy = []
rho_arr = []

for i, theta in enumerate(scan_arr):   
    dm = get_dm_w(theta) 
    rho_arr.append(qt.Qobj(dm.data,dims=[[2,2,2],[2,2,2]])) 
#     rho_arr_noisy.append(apply_u( qt.Qobj(dm.data,dims=[[2,2,2],[2,2,2]]))) 
    
c_01 = []
n_01 = []

c_02 = []
n_02 = []

c_12 = []
n_12 = []


for state in rho_arr:
    c_01.append(qt.entropy_vn(state.ptrace([0])))
    n_01.append(qt.concurrence(state.ptrace([0,1])))
    
    c_02.append(qt.entropy_vn(state.ptrace([1])))
    n_02.append(qt.concurrence(state.ptrace([0,2])))
    
    c_12.append(qt.entropy_vn(state.ptrace([2])))
    n_12.append(qt.concurrence(state.ptrace([1,2])))

fig, axs = plt.subplots(3,2)
axs[0,0].plot(c_01)
axs[0,1].plot(n_01)

axs[1,0].plot(c_02)
axs[1,1].plot(n_02)

axs[2,0].plot(c_12)
axs[2,1].plot(n_12)

plt.tight_layout()
```
